```python
monthly_kpis = (
    onlinesales
    .groupby(['Year', 'Month', 'Month_Name'])
    .agg(
        Monthly_Revenue=('Invoice_Value', 'sum'),
        Monthly_Orders=('Transaction_ID', 'nunique'),
        Monthly_Customers=('CustomerID', 'nunique')
    )
    .reset_index()
)
```

```python
monthly_kpis.head()
```

```python
category_kpis = (
    onlinesales
    .groupby('Product_Category')
    .agg(
        Category_Revenue=('Invoice_Value', 'sum'),
        Category_Quantity=('Quantity', 'sum'),
        Category_Orders=('Transaction_ID', 'nunique')
    )
    .reset_index()
    .sort_values(by='Category_Revenue', ascending=False)
)
```

```python
customer_features = (
    onlinesales
    .groupby('CustomerID')
    .agg(
        Total_Revenue=('Invoice_Value', 'sum'),
        Total_Orders=('Transaction_ID', 'nunique'),
        Total_Quantity=('Quantity', 'sum'),
        Avg_Order_Value=('Invoice_Value', 'mean'),
        First_Purchase_Date=('Transaction_Date', 'min'),
        Last_Purchase_Date=('Transaction_Date', 'max')
    )
    .reset_index()
)
```

```python
customer_features.head()
```

```python
reference_date = onlinesales['Transaction_Date'].max()
customer_features['Recency_Days'] = (reference_date - customer_features['Last_Purchase_Date']).dt.days
customer_features['Active_Days'] = (customer_features['Last_Purchase_Date'] - customer_features['First_Purchase_Date']).dt.days
```

```python
customer_features[['CustomerID','Recency_Days','Active_Days']].head()
```

```python
customer_ml = customer_features.merge(customers,on='CustomerID',how='left')
```

```python
customer_ml.head()
```

```python
customer_ml['Customer_Value_Segment'] = pd.qcut(customer_ml['Total_Revenue'],q=3,labels=['Low','Medium','High'])
```

```python
customer_ml['Customer_Value_Segment'].value_counts()
```

```python
y = customer_ml['Customer_Value_Segment']
```

```python
X = customer_ml.drop(columns=['CustomerID','First_Purchase_Date','Last_Purchase_Date','Customer_Value_Segment'])
```

```python
X.head()
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
log_reg = LogisticRegression(max_iter=1000,multi_class='multinomial')
```

```python
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
dt_model = DecisionTreeClassifier(max_depth=5,min_samples_split=20,random_state=42)
```
